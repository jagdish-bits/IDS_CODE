# 4 Recommender System For Baby Food

Recommendation system for baby food is implemented. A customer is prompted to select baby age from a dropdown menu. Baby Products
are recommended to the customer based on the rating, popularity and price of the product taking into consideration the age of the baby. Recommendations are given to the customer based on the following criteria: 
1. Products with the highest average rating across all online platforms are presented to the customer.
2. Products with the highest number of reviews are presented as most popular to the customer.
3. Products with the lowest price on any online platform are listed.

```python
TCOLOR = '\033[2;36m'
TENDC = '\033[m'



def recommended_product_by_rating(stage): 
     if stage != None :
        stage_index = result.loc[result['Stage'] == stage]
        means_stage_product = stage_index.groupby(['Company','Product',])['Rating'].mean()
       
        sort_product = means_stage_product.sort_values(ascending = False)
        print('===================================================')
        
        
        print(TCOLOR + 'Products with best ratings'+ TENDC)
        print(sort_product[:3])
       
#         print(sort_product)
     else :
        print('Invalid stage') 
```

```python
 def recommended_product_by_review(stage):

        stage_index_review = result.loc[result['Stage'] == stage]
        review_stage_product = stage_index_review.groupby(['Company','Product'])['Review'].sum()

        sort_product_review = review_stage_product.sort_values(ascending = False)
        print('===================================================')
        print(TCOLOR + 'Most Popular Products'+TENDC)
        print(sort_product_review[:3])
```

```python
def recommended_product_by_price(stage):   

        stage_index_price = result.loc[result['Stage'] == stage]
        price_stage_product = stage_index_price.groupby(['Company','Product','E_Comm'])['Price'].min()
        sort_price_product = price_stage_product.sort_values(ascending = True)
        print('===================================================')
        print(TCOLOR + 'Cheapest Product Pricing along with E_Commerce Platforms'+TENDC)
        print(sort_price_product[:5])
```
